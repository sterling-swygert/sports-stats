```python
import re

def clean_table(df: pd.DataFrame, post=False) -> pd.DataFrame:
    df = df.copy()
    if len(df.columns) > 0 and (type(df.columns[0]) != str) and len(df.columns[0]) > 1:
        new_cols = [(re.sub("Unnamed: [0-9]*_level_[0-9]*", "Misc", col[0]) + "_" + col[1]).rstrip('_') for col in df.columns if "Unnamed" not in col[1]]
        df.columns = df.columns.droplevel()
        idx = [i for i, col in enumerate(df.columns) if "Unnamed" not in col]
        df = df.iloc[:, idx]
        df.columns = [col.replace(" ", "-") for col in new_cols]
    if post:
        if "Year" in df.columns:
            df = df[df.Year.apply(lambda x: type(to_int(x)) == int)].reset_index(inplace=False).drop(columns=["index"])
        elif "Misc_Result" in df.columns:
            df = df[df.Misc_Result.apply(lambda x: x[0] if (len(x) > 1 and x[:2] in ["W ", "L ", "T "]) else 'NA')].reset_index(inplace=False).drop(columns=["index"])
    return df
```

```python
def get_team_links():
    headers = scraper.headers
    page_count = 1
    page_num = 1
    teams_url = 'https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/teams?page={}'
    links = []
    while page_num <= page_count:
        response = requests.get(url=teams_url.format(str(page_num)))
        resp = response.json()
        page_count = resp.get('pageCount', 1)
        page_num += 1
        for item in resp.get('items'):
            links.append(item.get('$ref'))
    return links

def get_athlete_links():
    headers = scraper.headers
    page_count = 1
    page_num = 1
    athletes_url = 'https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/athletes?limit=1000&page={}'
    links = []
    while page_num <= page_count:
        response = requests.get(url=athletes_url.format(str(page_num)))
        resp = response.json()
        page_count = resp.get('pageCount', 1)
        page_num += 1
        for item in resp.get('items'):
            links.append(item.get('$ref'))
    return links
    
```

```python
from requests import HTTPError

def get_team_data(ref):
    data = dict()
    response = requests.get(url=ref, headers=scraper.headers)
    if response.status_code == 200:
        resp = response.json()
        data['team_id'] = resp.get('id')
        data['team_uid'] = resp.get('uid')
        data['team_guid'] = resp.get('guid')
        data['location'] = resp.get('location')
        data['name'] = resp.get('name')
        data['abbrev'] = resp.get('abbreviation')
        data['active'] = resp.get('isActive')
        return data
    else:
        raise HTTPError(response.message)

def get_player_data(ref):
    data = dict()
    response = requests.get(url=ref, headers=scraper.headers)
    if response.status_code == 200: 
        resp = response.json()
        data['player_id'] = resp.get('id')
        data['player_uid'] = resp.get('uid')
        data['player_guid'] = resp.get('guid')
        data['first_name'] = resp.get('firstName')
        data['last_name'] = resp.get('lastName')
        data['height'] = resp.get('height')
        data['weight'] = resp.get('weight')
        data['birthtime'] = resp.get('dateOfBirth')
        draft = resp.get('draft')
        position = resp.get('position')
        college = resp.get('college')
        exp = resp.get('experience')
        birthplace = resp.get('birthplace')
        if position:
            data['position_id'] = position.get('id')
            data['position'] = position.get('abbreviation')
        if college:
            data['college_id'] = college.get('$ref').split('colleges/')[-1].split('?')[0]
        if exp:
            data['years_exp'] = exp.get('years', 0)
        if draft:
            data['draft_round'] = draft.get('round', 0)
            data['draft_pick'] = draft.get('selection', 0)
            data['draft_year'] = draft.get('year', 0)
        if birthplace:
            data['birth_city'] = birthplace.get('city', '')
            data['birth_state'] = birthplace.get('state', '')
            data['birth_country'] = birthplace.get('country', '')
        data['active'] = resp.get('status').get('id') == 1
        return data
    else:
        raise HTTPError(f'Error in request for {ref}\n: {response.message}')
```
